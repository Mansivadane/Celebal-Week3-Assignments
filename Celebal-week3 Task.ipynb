List

```python
#List
l=[] # empty list
print(l)
print(type(l))
```

```python
l=[1,2,6,9,2,5] #list having integer values
ls=['sanjivani', 'celebal', 'students'] # list having string values
lst=[1,3,5,'python',4.5,9] #list with mixed values
print(l)
print(ls)
print(lst)
```

```python
my_list=['c','e','l','e','b','a','l']
print(my_list)

#access 0th element
print(my_list[0])

#access middle element
print(my_list[4])

#access last element
print(my_list[-1])

# Nested List
n_list = ["Happy", [2, 0, 1, 5]]

# Nested indexing
print(n_list[0][1])

print(n_list[1][3])
```

```python
# List slicing in Python

my_list = ['p','r','o','g','r','a','m','e','r']

# elements from index 2 to index 4
print(my_list[2:5])

# elements from index 5 to end
print(my_list[5:])

# elements beginning to end
print(my_list[:])
```

Tuple

```python
# empty tuple
t=()
print(t)
print(type(t))
```

```python
# Tuple having integers
my_tuple = (4,5,9,1)
print(my_tuple)

# tuple with mixed datatypes
my_tuple1 = (1, "Hello", 3.4)
print(my_tuple)

# nested tuple
my_tuple = ("student", [4,7,9], (1, 2, 3))
print(my_tuple)

#tuple indexing
print(my_tuple[2][2])

my_tup=('c','e','l','e','b','a','l')
print(my_tup[5])

#negative indexing
print(my_tup[-1])
```

```python
# Accessing tuple elements using slicing
my_tuple = ('c','e','l','e','b','a','l')

# elements beginning to end
print(my_tuple[:])

# elements 2nd to 4th
print(my_tuple[1:4])

# elements beginning to 2nd
print(my_tuple[:-6])

# elements 8th to end
print(my_tuple[6:])
```

```python
# changing tuple values
my_tuple = (2,5,8,9,1)
print(my_tuple)

my_tuple = ('c','e','l','e','b','a','l')
print(my_tuple)
```

Set

```python
# creating empty set
s = set()
print(type(s))
```

```python
# set with nested values
my_set = {1,3,5,'mansi', 'gaurav', 3.4, 5.9}
print(my_set)

#add elements to set
my_set.add(12)
for i in my_set:
  print(i)

#update set
my_set.update([22,'shraddha'])
print(my_set)

#delete element from set
my_set.discard(3)
my_set.remove('gaurav')
print(my_set)
```

Dictionary

```python
# empty dictionary
my_dict = {}
print(type(my_dict))
```

```python
# dictionary with integer keys
my_dict = {1: 'Mansi', 2: 'Python'}
print(my_dict)

# dictionary with mixed keys
my_dict = {'name': 'Sanjivani', 1: [2, 4, 3]}
print(my_dict)

# using dict()
my_dict = dict({1:'Comp', 2:'IT'})
print(my_dict)

# from sequence having each item as a pair
my_dict = dict([('name','sanjivani'), (2,'COEP')])
print(my_dict)

#access the values from dictionary
print(my_dict['name'])
print(my_dict.get(2))
```

Membership Operators

```python
# In Operator
x = 5
if (type(x) is int):
    print("true")
else:
    print("false")

str= "Celebal"
print ('le' in str)
print ('lcb' in str)

#Not In Operator
x = 5.2
if (type(x) is not int):
    print("true")
else:
    print("false") 
    
str1= "Sanjivani"
print('sanjivani' not in str1)
print('celebal' not in str1)
```
